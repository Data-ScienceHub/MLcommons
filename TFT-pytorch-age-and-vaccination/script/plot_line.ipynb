```python
# Import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Import data
data_019 = pd.read_csv('../results/age_groups_019/figures_morris/Morris_019.csv')
data_2029 = pd.read_csv('../results/age_groups_2029/figures_morris/Morris_2029.csv')
data_3039 = pd.read_csv('../results/age_groups_3039/figures_morris/Morris_3039.csv')
data_4049 = pd.read_csv('../results/age_groups_4049/figures_morris/Morris_4049.csv')
data_5064 = pd.read_csv('../results/age_groups_5064/figures_morris/Morris_5064.csv')
data_6579 = pd.read_csv('../results/age_groups_6579/figures_morris/Morris_6579.csv')
data_80PLUS = pd.read_csv('../results/age_groups_80PLUS/figures_morris/Morris_80PLUS.csv')
```

```python
# Age 
AGE019 = data_019[data_019.Feature == 'AGE019']
AGE2029 = data_2029[data_2029.Feature == 'AGE2029']
AGE3039 = data_3039[data_3039.Feature == 'AGE3039']
AGE4049 = data_4049[data_4049.Feature == 'AGE4049']
AGE5064 = data_5064[data_5064.Feature == 'AGE5064']
AGE6579 = data_6579[data_6579.Feature == 'AGE6579']
AGE80PLUS = data_80PLUS[data_80PLUS.Feature == 'AGE80PLUS']

# Vaccination
Vaccination019 = data_019[data_019.Feature == 'VaccinationFull']
Vaccination2029 = data_2029[data_2029.Feature == 'VaccinationFull']
Vaccination3039 = data_3039[data_3039.Feature == 'VaccinationFull']
Vaccination4049 = data_4049[data_4049.Feature == 'VaccinationFull']
Vaccination5064 = data_5064[data_5064.Feature == 'VaccinationFull']
Vaccination6579 = data_6579[data_6579.Feature == 'VaccinationFull']
Vaccination80PLUS = data_80PLUS[data_80PLUS.Feature == 'VaccinationFull']
```

```python
# Plot
import matplotlib as mpl
mpl.rcParams['figure.dpi'] = 2000

fig, axes = plt.subplots(nrows=1, ncols=2, sharex=True, sharey=True, figsize=(7, 3.5))

axes[0].plot(AGE019.Delta, AGE019.Morris_sensitivity, label='0-19', color='black', linestyle=':')
axes[0].plot(AGE2029.Delta, AGE2029.Morris_sensitivity, label='20-29', color='red')
axes[0].plot(AGE3039.Delta, AGE3039.Morris_sensitivity, label='30-39', color='purple')
axes[0].plot(AGE4049.Delta, AGE4049.Morris_sensitivity, label='40-49', color='green')
axes[0].plot(AGE5064.Delta, AGE5064.Morris_sensitivity, label='50-64', color='brown')
axes[0].plot(AGE6579.Delta, AGE6579.Morris_sensitivity, label='65-79', color='grey')
axes[0].plot(AGE80PLUS.Delta, AGE80PLUS.Morris_sensitivity, label='80+', color='grey', linestyle=':')

axes[0].set_xlabel('Delta')
axes[0].set_ylabel('Scaled Morris Index')
axes[0].set_title('Age')
axes[0].legend(loc='upper left')

axes[1].plot(Vaccination019.Delta, Vaccination019.Morris_sensitivity, label='0-19', color='black', linestyle=':')
axes[1].plot(Vaccination2029.Delta, Vaccination2029.Morris_sensitivity, label='20-29', color='red')
axes[1].plot(Vaccination3039.Delta, Vaccination3039.Morris_sensitivity, label='30-39', color='purple')
axes[1].plot(Vaccination4049.Delta, Vaccination4049.Morris_sensitivity, label='40-49', color='green')
axes[1].plot(Vaccination5064.Delta, Vaccination5064.Morris_sensitivity, label='50-64', color='brown')
axes[1].plot(Vaccination6579.Delta, Vaccination6579.Morris_sensitivity, label='65-79', color='grey')
axes[1].plot(Vaccination80PLUS.Delta, Vaccination80PLUS.Morris_sensitivity, label='80+', color='grey', linestyle=':')

axes[1].set_xlabel('Delta')
axes[1].set_title('Vaccination')
plt.subplots_adjust(wspace=0, hspace=0)
plt.savefig('results.png')
plt.show()
```
