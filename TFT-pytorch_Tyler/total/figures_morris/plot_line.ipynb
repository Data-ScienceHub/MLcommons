```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('Morris.csv')
```

```python
AgeDist = data[data.Feature == 'AgeDist']
HealthDisp = data[data.Feature == 'HealthDisp']
DiseaseSpread = data[data.Feature == 'DiseaseSpread']
Transmission = data[data.Feature == 'Transmission']
VaccinationFull = data[data.Feature == 'VaccinationFull']
SocialDist = data[data.Feature == 'SocialDist']
```

```python
import matplotlib as mpl
mpl.rcParams['figure.dpi'] = 200


x =np.array([0, 0.0001,0.001,0.01,0.1,])

plt.semilogx(AgeDist.Delta, AgeDist.Morris_sensitivity, label='AgeDist', color='darkred')
plt.semilogx(VaccinationFull.Delta, VaccinationFull.Morris_sensitivity, label='VaccinationFull', color='darkorange')
plt.semilogx(HealthDisp.Delta, HealthDisp.Morris_sensitivity, label='HealthDisp', color='gold')
plt.semilogx(SocialDist.Delta, SocialDist.Morris_sensitivity, label='SocialDist', color='darkgreen')
plt.semilogx(Transmission.Delta, Transmission.Morris_sensitivity, label='Transmission', color='navy')
plt.semilogx(DiseaseSpread.Delta, DiseaseSpread.Morris_sensitivity, label='DiseaseSpread', color='rebeccapurple')

plt.ylim([0, 0.000035])
plt.ylabel('Scaled Morris Index')
plt.xlabel('Delta')
#plt.grid(True)

plt.legend()
```
